```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
spotify_df = pd.read_csv("top10s.csv", encoding = "ISO-8859-1")
spotify_df.head()
spotify_big = pd.read_csv("Spotify-2000.csv", encoding = "ISO-8859-1")
spotify_big.head()
```

```python
spotify_means = spotify_df.copy()
del spotify_means['title']
del spotify_means['artist']
del spotify_means['top genre']

spotify_means = spotify_means.groupby('year').mean()
spotify_means
```

Danceability: mennyire alkalmas az adott szám a táncolásra különböző zenei elemek alapján
Valence: mennyire pozitív az adott zene
Energy: mennyire intenzív, aktív az adott zene
Beats Per Minute (BPM): az adott zene tempója, zenei kifejezések alapján
Loudness: az adott zene átlagos hangereje decibelben
Speechiness: az elhangzott beszélsz szöveg az adott zenében
Liveness: az adott zene stúdióban lett rögzítve, vagy esetleg élőben egy koncerten
Acousticness: hangszerezése az adott zenének

```python
ax = spotify_means[['nrgy']].plot(kind='bar', title ="The average energy of the songs by year", figsize=(15, 10), legend=True, fontsize=12, color ="#1DB954" )
ax = spotify_means[['dnce']].plot(kind='bar', title ="The average danceability of the songs by year", figsize=(15, 10), legend=True, fontsize=12, color ="#1DB954" )
ax = spotify_means[['live']].plot(kind='bar', title ="The average liveness of the songs by year", figsize=(15, 10), legend=True, fontsize=12, color ="#1DB954" )
ax = spotify_means[['val']].plot(kind='bar', title ="The average valence of the songs by year", figsize=(15, 10), legend=True, fontsize=12, color ="#1DB954" )
ax = spotify_means[['spch']].plot(kind='bar', title ="The average speechiness of the songs by year", figsize=(15, 10), legend=True, fontsize=12, color ="#1DB954" )
ax = spotify_means[['acous']].plot(kind='bar', title ="The average acousticness of the songs by year", figsize=(15, 10), legend=True, fontsize=12, color ="#1DB954" )
```

A fenti oszlopdiagramokból a viszonylag logikus és egyértelmű adatok mellett számomra nemvárt változásokat véltem felfedezni.
Az egyre kevesebb élő felvétel (liveness) és a növekvő danceability nem volt meglepetés.
Ellenben a valance csökkenése mindenképp érdekesnek tűnik számomra, valamint kíváncsian várom, hogy a jelenlegi helyzet (korona-vírus) elülte után ez az érték vajon felfelé fog mozdulni, vagy esetleg nagyot csökken.
Az acousticness is meglepő fordulat volt számomra, úgy gondoltam, hogy az elektronikus zene az egyre modernebb eszközöknek köszönhetően felfelé ívelő pályát mutat be, de úgy néz ki a 2010-es évek végére a hangszerek kerültek az előtérbe.

```python
spotify_big_means = spotify_big.copy()
del spotify_big_means['Title']
del spotify_big_means['Artist']
del spotify_big_means['Top Genre']
del spotify_big_means['Length (Duration)']
del spotify_big_means['Index']
del spotify_big_means['Popularity']

spotify_big_means = spotify_big_means.groupby('Year').mean()
spotify_big_means
```

```python
pd.plotting.scatter_matrix(spotify_big_means, figsize=(15,15))
```

A mátrixunk néhány várt gondolatot alátámaszt, mit például az energy-loudness kapcsolatát, de számomra néhány meglepő eredményt is hozott.
Várható volt, hogy minél nagyobb az acousticness értéke, annál kisebb az Energy értéke, de nem számítottam rá, hogy ekkore mértékben,
A loudness-speechiness közötti korrelációt is meglepően tapasztaltam, mert egy sok szöveget tartalmazó zenéről nem feltétlenül az jut eszembe, hogy hangos.

```python
spotify_means_web = spotify_means.copy()
del spotify_means_web['Unnamed: 0']
del spotify_means_web['pop']
del spotify_means_web['dur']
spotify_means_web
```

```python
angles=np.linspace(0, 2*np.pi, len(spotify_means_web.columns), endpoint=False) 
angles=np.concatenate((angles,[angles[0]]))

y2010 = spotify_means_web.loc[2010]
s2010=np.concatenate((y2010,[y2010[0]]))
fig2010 = plt.figure()
ax2010 = fig2010.add_subplot(111, polar=True)
ax2010.plot(angles, s2010, "o-")
ax2010.fill(angles, s2010, alpha=0.25)
ax2010.set_thetagrids(angles * 180/np.pi, spotify_means_web.columns)    
plt.title('2010')
plt.show()

y2013 = spotify_means_web.loc[2013]
s2013=np.concatenate((y2013,[y2013[0]]))
fig2013 = plt.figure()
ax2013 = fig2013.add_subplot(111, polar=True)
ax2013.plot(angles, s2010, "o-")
ax2013.fill(angles, s2010, alpha=0.25)
ax2013.plot(angles, s2013, "o-", color = "#1DB954")
ax2013.fill(angles, s2013, "#1DB954", alpha=0.25)
ax2013.set_thetagrids(angles * 180/np.pi, spotify_means_web.columns)    
plt.title('2010 compared to 2013')
plt.show()

y2016 = spotify_means_web.loc[2016]
s2016=np.concatenate((y2016,[y2016[0]]))
fig2016 = plt.figure()
ax2016 = fig2016.add_subplot(111, polar=True)
ax2016.plot(angles, s2013, "o-", color = "#1DB954")
ax2016.fill(angles, s2013, "#1DB954", alpha=0.25)
ax2016.plot(angles, s2016, "o-", color = "#EC4747")
ax2016.fill(angles, s2016, "#EC4747",alpha=0.25)
ax2016.set_thetagrids(angles * 180/np.pi, spotify_means_web.columns)    
plt.title('2013 compared to 2016')
plt.show()

y2019 = spotify_means_web.loc[2019]
s2019=np.concatenate((y2019,[y2019[0]]))
fig2019 = plt.figure()
ax2019 = fig2019.add_subplot(111, polar=True)
ax2019.plot(angles, s2016, "o-", color = "#EC4747")
ax2019.fill(angles, s2016, "#EC4747",alpha=0.25)
ax2019.plot(angles, s2019, "o-", color = "#561F64")
ax2019.fill(angles, s2019, "#561F64",alpha=0.25)
ax2019.set_thetagrids(angles * 180/np.pi, spotify_means_web.columns)    
plt.title('2016 compared to 2019')
plt.show()

figdec = plt.figure()
axdec = figdec.add_subplot(111, polar=True)
axdec.plot(angles, s2010, "o-")
axdec.fill(angles, s2010, alpha=0.25)
axdec.plot(angles, s2019, "o-", color = "#561F64")
axdec.fill(angles, s2019, "#561F64",alpha=0.25)
axdec.set_thetagrids(angles * 180/np.pi, spotify_means_web.columns)    
plt.title('2010 compared to 2019')
plt.show()
```

Láthatóan az elmúlt évtizedben a trendek nem változtak akkora mértékben, hogy a következő néhány évben nagyobb változásokra számítsunk.

```python
y1959 = spotify_big_means.loc[1959]
s1959=np.concatenate((y1959,[y1959[0]]))
fig1959 = plt.figure()
ax1959 = fig1959.add_subplot(111, polar=True)

ax1959.plot(angles, s1959, "o-")
ax1959.fill(angles, s1959, alpha=0.25)

y1969 = spotify_big_means.loc[1969]
s1969=np.concatenate((y1969,[y1969[0]]))
ax1959.plot(angles, s1969, "o-")
ax1959.fill(angles, s1969, alpha=0.25)

y1979 = spotify_big_means.loc[1979]
s1979=np.concatenate((y1979,[y1979[0]]))
ax1959.plot(angles, s1979, "o-")
ax1959.fill(angles, s1979, alpha=0.25)

y1989 = spotify_big_means.loc[1989]
s1989=np.concatenate((y1989,[y1989[0]]))
ax1959.plot(angles, s1989, "o-")
ax1959.fill(angles, s1989, alpha=0.25)

y1999 = spotify_big_means.loc[1999]
s1999=np.concatenate((y1999,[y1999[0]]))
ax1959.plot(angles, s1999, "o-")
ax1959.fill(angles, s1999, alpha=0.25)

y2009 = spotify_big_means.loc[2009]
s2009=np.concatenate((y2009,[y2009[0]]))
ax1959.plot(angles, s2009, "o-")
ax1959.fill(angles, s2009, alpha=0.25)

y2019 = spotify_big_means.loc[2019]
s2019=np.concatenate((y2019,[y2019[0]]))
ax1959.plot(angles, s2019, "o-")
ax1959.fill(angles, s2019, alpha=0.25)

ax1959.set_thetagrids(angles * 180/np.pi, spotify_big_means.columns)    
plt.title('1959-2019')
plt.show()

figcent = plt.figure()
axcent = figcent.add_subplot(111, polar=True)
axcent.plot(angles, s1959, "o-")
axcent.fill(angles, s1959, alpha=0.25)
axcent.plot(angles, s2019, "o-")
axcent.fill(angles, s2019, alpha=0.25)
axcent.set_thetagrids(angles * 180/np.pi, spotify_big_means.columns)    
plt.title('1959 compared to 2019')
plt.show()
```

1959-2019 között sincs óriási változás néhány értékben.
Az acousticness elkönyvelhető volt előre, hogy még így is, hogy az elmúlt években újra növekedésnek indult, még mindig jóval kisebb értéket mutat, mint 1959-ben.
A BPM érték számomra meglepő volt, ugyanis úgy gondoltam, az elektronikus zenék ebben lekörözik a régi felvételeket.
A valance csökkenése érdekes, lehet társadalmi szempontból érdemes vizsgálni ezt a visszaesést.
